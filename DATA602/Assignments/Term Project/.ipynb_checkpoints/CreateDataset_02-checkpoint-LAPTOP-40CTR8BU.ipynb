```python
import pandas as pd
import yfinance as yf

msft = yf.Ticker("MSFT")

hist_msft = msft.history(period="20y", interval="1D", auto_adjust=False, prepost=True)
hist_msft.info()
```

```python
import copy
df = copy.deepcopy(hist_msft)
```

#### Volume Indicators
- AD (Chaikin Accumulation/Distribution line) {https://www.fmlabs.com/reference/default.htm?url=AccumDist.htm}

```python
import talib

df["AD"] = talib.AD(df["High"], df["Low"], df["Close"], df["Volume"])
df.plot(y="AD", grid=True, figsize=(10,10))
```

#### Momentum Indicators
- ADX    (Absolute Directional Movement Index)
- APO    (Absolute Price Oscillator)
- MFI    (Money Flow Index)
- WILLR  (Williams %R)
- CMO    (Chande Momentum Oscillator)
- PPO    (Percentage Price Oscillator)
- MACD   (Moving Average Convergence / Divergence)
- DX     (Directional Movement Index)
- RSI    (Relative Strength Index
- STOCH  (Stochastic)

```python
#AbsoluteDirectional Movement Index
df["ADX"] = talib.ADX(df["High"], df["Low"], df["Close"], timeperiod=15)

# Absolute price Oscillator
df["APO"] = talib.APO(df["Close"], fastperiod=12, slowperiod=26, matype=0)

# Money FLow Index
df["MFI"] = talib.MFI(df["High"], df["Low"], df["Close"], df["Volume"], timeperiod=15)

# Williams %R
df["WILLR"] = talib.WILLR(df["High"], df["Low"], df["Close"], timeperiod=15)

# Chande Momentum Oscillator
df["CMO"] = talib.CMO(df["Close"], timeperiod=15)

# Percent Price Oscillator
df["PPO"] = talib.PPO(df["Close"], fastperiod=12, slowperiod=26, matype=0)

# Moving Average Convergence/Divergence
df["MACD"], df["MACDSIGNAL"], df["MACDHIST"] = talib.MACD(df["Close"], fastperiod=12, slowperiod=26, signalperiod=9)

# Directional Movement Index
df["DX"] = talib.DX(df["High"], df["Low"], df["Close"], timeperiod=15)

# Relative Strength INDEX
df["RSI"] = talib.RSI(df["Close"], timeperiod=15)

#Stochastic
df["STOCH (slowk)"], df["STOCH (slowd)"] = talib.STOCH(df["High"], df["Low"], df["Close"],
                                                       fastk_period=5, slowk_period=3, slowk_matype=0, slowd_period=3, slowd_matype=0)

df.plot(y=["ADX", "MFI", "APO", "WILLR", "CMO", "PPO", "MACD", "MACDSIGNAL", "MACDHIST", "RSI", "DX", "STOCH (slowk)", "STOCH (slowd)"], subplots=True, grid=True, figsize=(10,30))
```

#### Volatility Indicators
- ATR (Average True Range)

```python
df["ATR"] = talib.ATR(df["High"], df["Low"], df["Close"], timeperiod=15)
df.plot(y="ATR", figsize=(10,10), grid=True)
```

#### Statistic Functions
- BETA {https://en.wikipedia.org/wiki/Beta_(finance)}

```python
df["BETA"] = talib.BETA(df["High"], df["Low"], timeperiod=5)
df.plot(y="BETA", figsize=(10,10), grid=True)
```

#### Price Transform
- MEDPRICE (Median Price)

```python
df["MEDPRICE"] = talib.MEDPRICE(df["High"], df["Low"])

df.plot(y="MEDPRICE", figsize=(10,10), grid=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import backtrader as bt
import numpy as np

def GoldenCross(Close, fastperiod=15, slowperiod=30):

    close = copy.deepcopy(Close)
    sma_short = talib.SMA(close, timeperiod=7)
    sma_long = talib.SMA(close, timeperiod=21)

    test = np.zeros(len(sma_short))

    crossdown = np.empty(len(sma_short))
    crossdown[:] = np.NaN

    crossup = np.empty(len(sma_short))
    crossup[:] = np.NaN

    for i in range(0, len(sma_short)):
        if i == 0: continue
        if sma_short[i] >= sma_long[i]:
            test[i] = 1
        if sma_short[i] <= sma_long[i]:
            test[i] = -1
        if test[i] < test[i-1]:
            crossdown[i] = 1
        if test[i] > test[i-1]:
            crossup[i] = 1
            
    return crossup, crossdown

df["SMA15"] = talib.SMA(df["Adj Close"], timeperiod=15)
df["SMA30"] = talib.SMA(df["Adj Close"], timeperiod=30)

df["SMAResidual"] = (df["SMA15"] - df["SMA30"])
df["CrossUp"], df["CrossDown"] = GoldenCross(df["Adj Close"])

df['CrossUp'] = df['CrossUp'].replace(np.nan, 0)
df['CrossDown'] = df['CrossDown'].replace(np.nan, 0)

df["Cross"] = df["CrossUp"]+(-df["CrossDown"])
df["Cross"] = df["Cross"].apply(lambda x: 0 if x == np.nan else x)
df["Target"] = df["Cross"].shift(periods=-7)

data = copy.deepcopy(df.drop(columns=["Open", "High", "Close", "Low", "CrossUp", "CrossDown", "Cross", "MEDPRICE"]))
data = data[data["Target"].notna()]
data = data[data["SMA30"].notna()]
data = data[data["MACD"].notna()]
data.iloc[-20:]
```

```python
plt.figure(figsize=(15,20))
sns.heatmap(data.corr())
plt.show()
```
